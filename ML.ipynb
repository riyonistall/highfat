```python
def analyze_numbers(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    total = sum(numbers)
    avg = total / len(numbers)
    return min_val, max_val, avg, total
def main():
    nums = input("Enter numbers separated by space: ")
    nums_list = [float(x) for x in nums.split()]
    min_val, max_val, avg, total = analyze_numbers(nums_list)
    print("Minimum:", min_val)
    print("Maximum:", max_val)
    print("Average:", avg)
    print("Sum:", total)
main()
```

```python
def count_labels(labels):
    label_count = {}
    for label in labels:
        if label in label_count:
            label_count[label] += 1
        else:
            label_count[label] = 1
    return label_count

def main():
    labels = ['spam', 'ham', 'spam', 'spam', 'ham']
    result = count_labels(labels)
    print(result)
main()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load dataset
df = pd.read_csv('Heart.csv')

print("Shape of data :", df.shape)
print("\nMissing values :\n", df.isnull().sum())
print("\nData type :\n", df.dtypes)
print("\nCount of zeros in each column :\n", (df == 0).sum())
print("\nMean age of patients :", df['Age'].mean())

# Selecting required columns
selected_columns = ['Age', 'Sex', 'ChestPain', 'RestBP', 'Chol']
df_selected = df[selected_columns]

# Train-test split
main_data, test_data = train_test_split(df_selected, test_size=0.25, random_state=42)
print("\nTraining set shape :", main_data.shape)
print("Testing set shape :", test_data.shape)

# Confusion matrix values (example given)
TP = 45
FN = 5
FP = 65
TN = 395

conf_matrix = np.array([[TP, FP],
                        [FN, TN]])

print("\nConfusion Matrix :")
print(pd.DataFrame(conf_matrix,
                   columns=["Predicted Positive", "Predicted Negative"],
                   index=["Actual Positive", "Actual Negative"]))

# Performance metrics
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * (precision * recall) / (precision + recall)

print("\nPerformance Metrics")
print("Accuracy :", round(accuracy, 2))
print("Precision :", round(precision, 2))
print("Recall :", round(recall, 2))
print("F1 Score :", round(f1, 2))
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_curve, roc_auc_score
import matplotlib.pyplot as plt


try:
    df = pd.read_csv('Adm_Pred.csv')
except FileNotFoundError:
    print("Error: 'Adm_Pred.csv' not found. Please upload the file or provide the correct path.")
    # Exit or handle the error appropriately
    exit()

# Clean column names by stripping whitespace
df.columns = df.columns.str.strip()

# Assuming 'Chance of Admit' is the target variable
# Convert 'Chance of Admit' to a binary outcome (0 or 1)
if 'Chance of Admit' in df.columns:
    df['Chance of Admit'] = df['Chance of Admit'].apply(lambda x: 1 if x >= 0.5 else 0)
else:
    print("Error: 'Chance of Admit' column not found.")
    exit()

# Define features (X) and target (y)
# Assuming all other columns are features
X = df.drop(columns=['Chance of Admit'])
y = df['Chance of Admit']

# Define splits
splits = {
    "80-20": 0.2,
    "70-30": 0.3,
    "65-35": 0.35
}

accuracies = {}

plt.figure(figsize=(10, 6))

for label, test_size in splits.items():
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)

    # For binary classification, predict probabilities and then convert to binary predictions
    y_prob = model.predict(X_test)
    y_pred = [1 if p >= 0.5 else 0 for p in y_prob]

    acc = accuracy_score(y_test, y_pred)
    accuracies[label] = acc

    print(f"In --- {label} Split ---")
    print("Accuracy : ", acc)

    cm = confusion_matrix(y_test, y_pred)
    print("\nConfusion Matrix:")
    print(pd.DataFrame(cm, columns=["Predicted Negative", "Predicted Positive"], index=["Actual Negative", "Actual Positive"]))


    p = precision_score(y_test, y_pred)
    r = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    print("Precision:", p)
    print("Recall:", r)
    print("F1 Score:", f1)

    # Calculate ROC curve and AUC
    # roc_curve requires true binary labels and prediction probabilities
    # LinearRegression predict gives continuous values, which we can use as probabilities here
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    auc = roc_auc_score(y_test, y_prob)

    plt.plot(fpr, tpr, label=f'{label} (AUC = {auc:.3f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.show()

# Bar chart for accuracy comparison
plt.figure(figsize=(8, 5))
plt.bar(accuracies.keys(), accuracies.values(), color=['skyblue', 'salmon', 'lightgreen'])
plt.title("Accuracy Comparison for LR")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression # Changed from LinearRegression for predict_proba
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler # Added import for StandardScaler

try:
    df = pd.read_csv("Adm_Pred.csv")
except FileNotFoundError:
    print("Error: 'Adm_Pred.csv' not found. Please upload the file or provide the correct path.")
    exit()

df.columns = df.columns.str.strip()

# Assuming 'Chance of Admit' is the target variable
if 'Chance of Admit' in df.columns:
    df['Chance of Admit'] = df['Chance of Admit'].apply(lambda x: 1 if x >= 0.5 else 0)
else:
    print("Error: 'Chance of Admit' column not found.")
    exit()

x = df.drop(columns=['Chance of Admit'])
y = df['Chance of Admit']

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

splits = {'80-20': 0.2, '70-30': 0.3, '65-35': 0.35}
accuracies = {}

plt.figure(figsize=(10, 6))

for label, test_size in splits.items():
    x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=test_size, random_state=42)

    model = LogisticRegression() # Using Logistic Regression as it has predict_proba
    model.fit(x_train, y_train)

    y_prob = model.predict_proba(x_test)[:, 1] # Get probabilities for the positive class
    y_pred = model.predict(x_test)

    acc = accuracy_score(y_test, y_pred)
    accuracies[label] = acc

    print(f'In {label} Split:')
    print("Accuracy:", acc)
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:\n", cm)
    prec = precision_score(y_test, y_pred)
    r1 = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    print("Precision:", prec)
    print("Recall:", r1)
    print("F1 Score:", f1)

    fpr, tpr, _ = roc_curve(y_test, y_prob)
    auc = roc_auc_score(y_test, y_prob) # Corrected AUC calculation

    plt.plot(fpr, tpr, label=f'{label} (AUC = {auc:.3f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
plt.bar(accuracies.keys(), accuracies.values())
#color = ['steel blue', 'dark orange', 'medium sea green'] # This line is not used
plt.title("Accuracy Comparison for LR")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```
